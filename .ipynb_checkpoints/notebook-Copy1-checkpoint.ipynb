#### Solution Q1.e)

```python
from scipy.optimize import bisect

######################
#   SOLUTION Q1.e)   #
######################

def coef_trfunc(poles, zeros):
    # =========================================================================
    # COEF_TRFUNC FUNCTION
    #
    # ARGUMENTS:
    #    * poles: poles of a second-order filter
    #    * zeros: zeros of a second-order filter
    #
    # RETURN:
    #    * b: vector of coefficients of the filter transfer func. numerator
    #    * a: vector of coefficients of the filter transfer func. denominator
    #
    # SUMMARY:
    #    Find the transfer function coefficients of a second-order filter from
    #    its poles and zeros.
    # =========================================================================

    b = [1, -np.real((zeros[0] + zeros[1])), np.real(zeros[0] * zeros[1])]
    a = [1, -np.real((poles[0] + poles[1])), np.real(poles[0] * poles[1])]
    
    K_inv = np.max([np.abs(eval_trfunc(b, a, -1)), 
                    np.abs(eval_trfunc(b, a, 1))])

    if K_inv > 0.0:
        b = b/K_inv
    
    return b, a


def eval_trfunc(b, a, z):
    # =========================================================================
    # EVAL_TRFUNC FUNCTION
    #
    # ARGUMENTS:
    #    * b: vector of coefficients of the filter transfer func. numerator
    #    * a: vector of coefficients of the filter transfer func. denominator
    #    * z: values (array or single value) at which the transfer func. must
    #         be evaluated.
    #
    # RETURN:
    #    * The evaluation of the transfer func. at z
    #
    # SUMMARY:
    #    Evaluate a transfer function defined by its coefficients b and a at z
    # =========================================================================

    num = np.zeros(np.shape(z), dtype=np.complex128)
    den = np.zeros(np.shape(z), dtype=np.complex128)
    for i, b_val in enumerate(b):
        num += b_val * z**(-i)
    for i, a_val in enumerate(a):
        den += a_val * z**(-i)
    return num/den


def design_notch(fs, f_cut, bw_3db):
    # =========================================================================
    # DESIGN_NOTCH FUNCTION
    #
    # ARGUMENTS:
    #    * fs: sampling frequency
    #    * f_cut: frequency to cut off expressed in Hz
    #    * bw_3db: 3dB-bandwidth expressed in Hz
    #
    # RETURN:
    #    * b: vector of coefficients of the designed filter numerator
    #    * a: vector of coefficients of the designed filter denominator
    #    * poles: poles of the designed filter
    #    * zeros: zeros of the designed filter
    #
    # SUMMARY:
    #    Design a second-order notch filter for f_cut frequency with a bw_3db
    #    3dB-bandwidth
    # =========================================================================

    omega0 = 2 * np.pi * f_cut/fs
    zeros = [np.exp(1j * omega0), np.exp(-1j * omega0)]
    bw_up_lim = np.exp(1j * (omega0 + 2 * np.pi * bw_3db/(2*fs)))

    # Finding root of power gives the value of r that satisfies the 3dB bw
    def power_at_3db(r):
        poles = [r*np.exp(1j * omega0), r*np.exp(-1j * omega0)]
        b, a = coef_trfunc(poles, zeros)
        pow_up_lim = np.abs(eval_trfunc(b, a, bw_up_lim))**2
        return pow_up_lim - 0.5

    r = bisect(power_at_3db, 0.001, 0.999)
    poles = [r*np.exp(1j * omega0), r*np.exp(-1j * omega0)]
    b, a = coef_trfunc(poles, zeros)
    return b, a, zeros, poles
    

# Filter specifications
f_cut = 78 # Hz (deduced from spectrogram)
bw_3db = 5 # Hz

# Design notch filter
b, a, zeros, poles = design_notch(fs, f_cut, bw_3db)

# Pole-zero plot
omega = np.arange(0, 2*np.pi, 0.001)
unit_circle = np.exp(1j * omega)

fig, ax = plt.subplots(figsize=(7,7))
ax.scatter(np.real(poles), np.imag(poles), s=100, c='r', marker="x")
ax.scatter(np.real(zeros), np.imag(zeros), s=100, facecolors='none', edgecolors='b')
ax.plot(np.real(unit_circle), np.imag(unit_circle), 'k', markersize=0.5)
ax.set_xlim(-1.5, 1.5)
ax.set_ylim(-1.5, 1.5)
ax.set_xticks([-1, 1], [-1, 1])
ax.set_yticks([-1, 1], [-1, 1])
ax.spines['left'].set_position('center')
ax.spines['bottom'].set_position('center')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')
ax.set_title('Pole-zero plot')

# Bode plots
f = np.arange(0.0, 0.5, 0.001)
z = np.exp(1j * 2 * np.pi * f)
H = eval_trfunc(b, a, z)
pow_spec = np.abs(H)**2
phase = np.angle(H, deg=True)

fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(10,7), sharex=True)
axes[0].plot(f, 10*np.log10(pow_spec + 10**(-20)))
axes[0].set_title('Bode plots')
axes[0].set_ylabel('Magnitude [dB]')
axes[0].set_ylim(-50, 1)
axes[1].plot(f, phase)
axes[1].set_ylabel('Phase [°]')
axes[1].set_xlabel('Normalized frequency')
axes[1].set_ylim(-90, 90);
```
